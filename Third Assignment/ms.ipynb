#### "Merge Sort"

```python
def mer(A, B):
    C = []
    while len(A) > 0 and len(B)> 0:
        if A[0] > B[0]:
            C.append(B[0])
            B.pop(0)
        else:
            C.append(A[0])
            A.pop(0)
        if len(B) == 0:
            for e in A:
                C.append(e)
        if len(A) == 0:
            for e in B:
                C.append(e)
    return(C)
```
