#### "3-Way Partition"

```python
def threepar(A):
    mi = []
    eq = []
    maj = []
    for e in A:
        if e < A[0]:
            mi.append(e)
        elif e == A[0]:
            eq.append(e)
        else:
            maj.append(e)
    mi.extend(eq)
    mi.extend(maj)
    print(*mi)
```

```python
f = open('rosalind_par3.txt', 'r')
f = f.readlines()
n = int(f[0])
A = list(map(int, f[1].split()))
threepar(A)
```
