#### "Binary Search"

```python
A = list(map(int, input().split()))
M = list(map(int, input().split()))
def bins(A, M):
    opl = []
    for m in M:
        if m in A:
            opl.append(A.index(m)+1)
        else:
            opl.append(-1)
    return(opl)
print(*bins(A, M))
```
