#### "2-Way Partition"

```python
def par(A):
    B = []
    B.append(A[0])
    a = A[0]
    A.pop(0)
    for i in range(len(A)):
        if A[i] <= a:
            B.insert(0, A[i])
        elif A[i] > a:
            B.append(A[i])
    return(B)
```

```python
f = open('rosalind_par.txt', 'r')
f = f.readlines()
A = list(map(int, f[1].split()))
print(*par(A))
```
