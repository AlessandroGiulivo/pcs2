#### "Translating RNA into Protein"

```python
from Bio.Seq import Seq
```

```python
def prot(s):
    rna_seq = Seq(s)
    prot = rna_seq.translate(to_stop = True)
    return(prot)
```

```python
with open('rosalind_prot.txt', 'r') as f:
    f = f.read()
    s = str(f)
    print(prot(s))
```
